```python
selected_range = green_trip[(green_trip['lpep_pickup_datetime']>'2019-09-30')&(green_trip['lpep_dropoff_datetime']<'2019-11-01')]
selected_range
```

```python
def trip_segment(trip_distance):
    if trip_distance <= 1:
        return 'Up to 1 mile'
    elif trip_distance <= 3:
        return 'In between 1 (exclusive) and 3 miles (inclusive)'
    elif trip_distance <= 7:
        return 'In between 3 (exclusive) and 7 miles (inclusive)'
    elif trip_distance <= 10:
        return 'In between 7 (exclusive) and 10 miles (inclusive)'
    else:
        return 'Over 10 miles'  
    
    
selected_range['trip_segment'] = selected_range['trip_distance'].apply(trip_segment)
selected_range
```

```python
selected_range['trip_segment'].value_counts()
```
